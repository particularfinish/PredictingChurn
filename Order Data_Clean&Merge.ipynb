Order id's are not unique. Several records may be associated with the same order_id. We want the dataframe to be joined on *each* of those instances. Not just one of the instances. 

```python
# load cleaned df
df = pd.read_csv('data/orderskpi.csv')
df.info()

# Note in this dataset, order_id is unique. 
```

We need a many-to-one join. In df_items, there are unique values versus the total rows:

```python
len(df_items['order_id'].unique()), len(df_items)
```

In df, there is a unique order_id for each row:

```python
len(df['order_id'].unique().tolist()), len(df)
```

How many of those order_ids overlap?

```python
df1= pd.DataFrame((df_items.order_id.isin(df.order_id)))
```

```python
df1[df1.order_id == True].count()
```

```python
x = df_items['order_id'].unique().tolist()
y = df['order_id'].unique().tolist()

l_func = lambda x, y: list((set(x)- set(y))) + list((set(y)- set(x))) 


non_match = l_func(x, y)
```

```python
print(174973 - len(non_match))
```

The 'many' is in df_items, and the one is in df. We should get a dataframe of length 174973 as shown above. 

```python
merged = pd.merge(df_items, df, validate ="many_to_one")
```

    'number_of_purchases' - This ranges from 1 to 1302
